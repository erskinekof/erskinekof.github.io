````
from qiskit.circuit.library import ZZFeatureMap

num_features = train_features.shape[1]

feature_map = ZZFeatureMap(feature_dimension=num_features, reps=1)
feature_map.decompose().draw(output="mpl")
````

````
from qiskit.circuit.library import NLocal, RYGate, CXGate
from qiskit.circuit import QuantumCircuit

# Use the number of qubits from the feature map
num_qubits = feature_map.num_qubits

# Define a simple ansatz circuit using NLocal
# This is an example, you might need to adjust the gates and reps based on your problem
ansatz = NLocal(
    num_qubits,
    [RYGate(0)], # Pass rotation_gates positionally
    CXGate(),    # Pass entanglement_gates positionally
    reps=1,
    insert_barriers=True
)
ansatz.decompose().draw(output="mpl")
````

# Task
Tune the hyperparameters of the Random Forest model using GridSearchCV and evaluate the performance of the tuned model.

## Import necessary libraries

### Subtask:
Import necessary libraries for hyperparameter tuning of the Random Forest model.


**Reasoning**:
Import the necessary libraries for hyperparameter tuning of the Random Forest model.



````
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
````

## Define hyperparameter grid

### Subtask:
Specify a range of values for the hyperparameters you want to tune (e.g., `n_estimators`, `max_depth`, `min_samples_split`, `min_samples_leaf`).


**Reasoning**:
Define the parameter grid for hyperparameter tuning of the Random Forest model.



````
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
````

## Initialize gridsearchcv

### Subtask:
Create a `GridSearchCV` object with the Random Forest model, the hyperparameter grid, and the desired number of cross-validation folds.


**Reasoning**:
Create the Random Forest model and the GridSearchCV object as instructed.



````
rf_model = RandomForestClassifier(random_state=123)

grid_search = GridSearchCV(rf_model, param_grid, cv=5)
````

## Perform grid search

### Subtask:
Fit the `GridSearchCV` object to the training data (`train_features`, `train_labels`) to find the best combination of hyperparameters.


**Reasoning**:
Fit the GridSearchCV object to the training data to find the best hyperparameters.



````
grid_search.fit(train_features, train_labels)
````

## Get best parameters and best score

### Subtask:
Access the best hyperparameters found by `GridSearchCV` and the corresponding best cross-validation score.


**Reasoning**:
Access and display the best hyperparameters and the best cross-validation score from the fitted GridSearchCV object.



````
best_params = grid_search.best_params_
best_score = grid_search.best_score_

print("Best Hyperparameters:", best_params)
print("Best Cross-Validation Score:", best_score)
````

## Evaluate the tuned model

### Subtask:
Train a new Random Forest model using the best hyperparameters found by GridSearchCV and evaluate its performance on the test data (`test_features`, `test_labels`).


**Reasoning**:
Train a new Random Forest model using the best hyperparameters found by GridSearchCV and evaluate its performance on the test data.



````
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Create a new Random Forest Classifier instance with the best hyperparameters
tuned_rf_model = RandomForestClassifier(n_estimators=best_params['n_estimators'],
                                        max_depth=best_params['max_depth'],
                                        min_samples_split=best_params['min_samples_split'],
                                        min_samples_leaf=best_params['min_samples_leaf'],
                                        random_state=123)

# Train the model using the training data
tuned_rf_model.fit(train_features, train_labels)

# Make predictions on the test data
tuned_rf_predictions = tuned_rf_model.predict(test_features)

# Evaluate the model
tuned_rf_accuracy = accuracy_score(test_labels, tuned_rf_predictions)
print(f"Tuned Random Forest Model Accuracy: {tuned_rf_accuracy:.2f}")

print("\nTuned Random Forest Model Confusion Matrix:")
print(confusion_matrix(test_labels, tuned_rf_predictions))

# Optionally, display a more detailed classification report
print("\nTuned Random Forest Model Classification Report:")
print(classification_report(test_labels, tuned_rf_predictions))
````
