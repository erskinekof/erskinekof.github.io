### Quantum Support Vector Machine (QSVM)

```
from sklearn.decomposition import PCA

# Dimensionality reduction using PCA
n_qubits = 4  # Number of qubits to use
pca = PCA(n_components=n_qubits)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# Define the quantum device
dev = qml.device("default.qubit", wires=n_qubits)

# Define the quantum circuit for feature encoding
@qml.qnode(dev)
def quantum_circuit(inputs, weights):
    for i in range(len(inputs)):
        qml.RX(inputs[i], wires=i)
    for i in range(len(inputs)):
        qml.Rot(weights[i][0], weights[i][1], weights[i][2], wires=i)
    return [qml.expval(qml.PauliZ(i)) for i in range(len(inputs))]
```

```
# Define a function to encode classical data into quantum features
def encode_data_to_quantum(data, weights):
    quantum_features = []
    for sample in data:
        circuit_output = quantum_circuit(sample, weights)
        quantum_features.append(circuit_output)
    return np.array(quantum_features)

# Train a QSVM model
from sklearn.svm import SVC # Import SVC for the SVM model
def train_qsvm(X_train, y_train, X_test, y_test, initial_weights):
    # Encode training and testing data into quantum features
    X_train_quantum = encode_data_to_quantum(X_train, initial_weights)
    X_test_quantum = encode_data_to_quantum(X_test, initial_weights)

    # Train a classical SVM model using the quantum features
    svm_model = SVC(kernel='linear')
    svm_model.fit(X_train_quantum, y_train)

    # Calculate approximate hinge loss on training data
    y_pred_train = svm_model.decision_function(X_train_quantum)
    hinge_loss = np.mean(np.maximum(0, 1 - y_train * y_pred_train))

    # Evaluate the model
    accuracy = svm_model.score(X_test_quantum, y_test)

    return svm_model, accuracy, hinge_loss
```

```
# Generate random initial weights for the quantum circuit
np.random.seed(0)
initial_weights = np.random.randn(n_qubits, 3)  # Adjusted for the number of qubits

# Train the QSVM model using PCA-reduced data
svm_model, accuracy, loss = train_qsvm(X_train_pca, y_train, X_test_pca, y_test, initial_weights)

# Print the accuracy and loss of the trained QSVM model
print("Accuracy of the QSVM model: {:.2f}%".format(accuracy * 100))
print("Approximate Hinge Loss:", loss)
```

### Anomaly Detection using Isolation Forest

Since the dataset contains only one class, we can treat this as an anomaly detection problem to identify data points that deviate from the normal behavior. Isolation Forest is an effective algorithm for this task.

```
from sklearn.ensemble import IsolationForest

# Initialize the Isolation Forest model
# contamination is the proportion of outliers in the data set. It is an important parameter
# for adjusting the sensitivity of the model to anomalies. We will use auto for now.
iso_forest = IsolationForest(random_state=42, contamination='auto')

# Fit the model to the scaled training data
# Note: Isolation Forest is an unsupervised method, so we only use X_train_scaled
iso_forest.fit(X_train_scaled)

# Predict anomalies (-1 for outliers, 1 for inliers)
anomaly_predictions_train = iso_forest.predict(X_train_scaled)
anomaly_predictions_test = iso_forest.predict(X_test_scaled)

# Get anomaly scores (lower score indicates a higher likelihood of being an anomaly)
anomaly_scores_train = iso_forest.decision_function(X_train_scaled)
anomaly_scores_test = iso_forest.decision_function(X_test_scaled)

print("Anomaly detection complete.")
print(f"Number of anomalies detected in training set: {list(anomaly_predictions_train).count(-1)}")
print(f"Number of anomalies detected in test set: {list(anomaly_predictions_test).count(-1)}")

# Display the first few anomaly predictions and scores for the test set
print("\nFirst 10 anomaly predictions for test set:")
print(anomaly_predictions_test[:10])
print("\nFirst 10 anomaly scores for test set:")
print(anomaly_scores_test[:10])
```
